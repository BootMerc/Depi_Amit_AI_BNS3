```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4])
y=np.array([2,2.8,3.6,4.5])
```

```python
w=0
b=0
alpha=0.01
num_iteration=20
```

```python
sse_values=[]

for i in range(num_iteration):
    y_hat=w*x+b
    d_w=2/4*np.sum((y_hat-y)*x)
    d_b=2/4*np.sum((y_hat-y))
    
    w-=alpha*d_w
    b-=alpha*d_b
    
    sse=(np.sum((y_hat-y)**2))
    sse_values.append(sse)
    
    if(i+1)%20==0:
        print(f"iteration{i+1},SSE:{sse}")
    
    
    
```
