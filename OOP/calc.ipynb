```python
print("Welcome to the Simple Calculator!")

menu = """Select an operation:
    1. Addition (+)
    2. Subtraction (-)
    3. Multiplication (*)
    4. Division (/)
"""
prompt = "Enter your choice (1/2/3/4) or 'exit' to quit: "

class Calculator:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
    def add(self):
        return self.num1 + self.num2
    def sub(self):
        return self.num1 - self.num2
    def mul(self):
        return self.num1 * self.num2
    def div(self):
        return self.num1 / self.num2 

while True:
    print(menu)
    operation = input(prompt).strip()
    if operation.lower() == 'exit':
        print('Exiting the calculator. Goodbye!')
        break

    if operation not in {'1', '2', '3', '4'}:
        print('Wrong Operation!')
        continue

    try:
        num1 = float(input("Enter the first number: ").strip())
        num2 = float(input("Enter the second number: ").strip())
    except ValueError:
        print("Invalid number input. Please enter valid numeric values.")
        continue

    calc = Calculator(num1, num2)

    try:
        if operation == '1':
            result = calc.add()
            symbol = '+'
        elif operation == '2':
            result = calc.sub()
            symbol = '-'
        elif operation == '3':
            result = calc.mul()
            symbol = '*'
        else:
            if num2 == 0:
                print("Error: Cannot divide by zero.")
                continue
            result = calc.div()
            symbol = '/'

        print(f"Result: {num1} {symbol} {num2} = {result}")
    except ZeroDivisionError:
        print("Error: Cannot divide by zero.")
```
